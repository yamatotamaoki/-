```python
from sklearn.datasets import load_wine
```

```python
data = load_wine()
```

```python
x = data.data[:,[0,9]]
print(x)
```

```python
from sklearn.cluster import KMeans
```

```python
n_clusters = 3
```

```python
model = KMeans(n_clusters=n_clusters)
```

```python
pred = model.fit_predict(x)
```
